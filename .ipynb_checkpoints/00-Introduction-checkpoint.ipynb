# Leveraging Jazz Album Meta-Characteristics for Market-Value Prediction

## Motivation

When shopping for vinyl records, it is common that most of the records you find in the store will be ones that are unfamiliar to you. If you are in luck, the store will have a record player with which which you can test the records that have caught your eye to see if they are worth the price being charged by the store.

However, outside of checking the going rate online for each record that catches your eye, individuals are relegated to gauging the fairness of the price being charged by the store by means of their own experience, previous purchase history, and a feel for the contemporary going rate of records, and compare that to their own preferences for the attributes of the individual record (such as the recording artist, the label the record was released on, the genre of the record, and even the cover image of the record).

In practice, individuals construct their own models for picking out records, in which they factor in both their own preferences, as well as their own estimated model of the market value of a given record given its meta-characteristics (such as the recording artist, the releasing label, the genre, etc.), to determine whether or not they should pick out a record given its stated price, test it if possible, and ultimately purchase it. As such, a trip to the record store for an record store patron can be naively represented as the following optimization problem:

where:

- $u_i(x_j)$ is $i$'s per-album utility function

- The optimization problem is constrained by the budget constraint
    - The individual's money, $M_i$, which is equal to the amount the individual has budgeted for albums, $C_i$, minus the price of all purchased items $P(X)$, must be greater than or equal to zero


In the estimation problem outlined above, the key components that determine the utility that an individual obtains from purchasing some album $x_j$ are the individual's own preferences, represented by $t_i(x_j)$ and the estimated arbitrage opportunity they would gain from purchasing the album in the store as opposed to on the larger album market $\hat{V}_m(x_i) - V_s(x_i)$.

Clearly, the first component of the utility function is entirely up to the individual, as it is wholly determined by their individual tastes and preferences. In the definition above, the second component is also individual-specific, as it incorporates the individual's own estimation of the market-value of the album, $\hat{V_m}$, and compares this to the store-value of the album. Note that the estimated market-value of the record could be wildly divergent from the true market-value, $V_m(x_i)$, depending on the parametrization of their mental model evaluating the value of the album given its meta-attributes, represented by the vector $x_j$. As such, we see that in the optimization problem of purchasing records in a store, the element keeping the individual from being able to perfectly optimize the utility they gain from shopping is their own estimation of the market value of a given record based on it's meta-attributes


For this project, I will use data taken from Discogs.com, the largest online second-hand marketplace for physical music, to create a non-individualistic estimation of album market-value, $\hat{V_m}$ for Jazz albums. With this, I aim to identify the divergence of $V_m\ -\ \hat{V_m}$ and identify to what extent it is possible to predict the true market-value of an album, based only on its meta-attributes, using machine learning models.

As the data used in this study is obtained from Discogs.com, we will have access to features that would not usually be available for evaluation by a record store patron, such as the number of individuals who own that record (on the marketplace), or what the average rating for that record is (by users of the marketplace). As such, we will develop two sets of models to predict the market-value of a given record. 

The first set of models will be developed under what we will refer to as the 'Record Store Scenario', in which we train our models only on those features which are ostensibly available for evaluation by an individual shopping for albums in a record store (which we use as a catch-all term for a store which sells physical music). The second set of models will then use all the information made available by the Discogs.com platform in order to predict the market-value of a given Jazz album. By juxtaposing these two scenarios, we hope to identify not only to what extent it is possible to predict the market-value of a Jazz album using only its meta-characteristics, but also the possible advantage that the inclusion of marketplace platform features can have in this prediction task, with respect to the traditional record store case.
